```python
not_ex_age = not_excercise[not_excercise['대분류'] == '연령별'].copy()
not_ex_income = not_excercise[not_excercise['대분류'] == '소득별'].copy()
not_ex_age
not_ex_income
```

```python
not_ex_age.drop(columns='대분류', inplace=True)
not_ex_income.drop(columns='대분류', inplace=True)
not_ex_age
not_ex_income
```

```python
not_ex_age.set_index('분류', inplace=True)
not_ex_income.set_index('분류', inplace=True)
not_ex_age
not_ex_income
```

```python
f, ax = plt.subplots(1, 2, figsize = (16, 8))
not_ex_age['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02, 0.02, 0.02, 0.02, 0.02], ax=ax[0], autopct='%1.1f%%')
not_ex_income['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02, 0.02, 0.02, 0.02, 0.02], ax=ax[1], autopct='%1.1f%%')
ax[0].set_title('연령별 운동할 시간이 없음')
ax[1].set_title('소득별 운동할 사람이 없음')
plt.show()
```
